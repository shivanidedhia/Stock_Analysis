```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
df.head(10)
```

## Graph 1

```python
ax1 = sns.catplot(x='StockTicker',y='Highest_Hourly_Price',data=df, kind='bar',palette="Set3")
ax1.fig.set_size_inches(12,10)
```

## Highest Hourly Price was held by stock ticker SHOP 

```python
SHOP = df[(df['StockTicker'] == "SHOP")]
box_company = sns.boxplot(x="StockTicker", y="Highest_Hourly_Price", data=SHOP, palette = "Set3")
ax1.fig.set_size_inches(12,10)
```

```python
sns.relplot(x="StockTicker", y="Highest_Hourly_Price", hue="StockTicker", data=df);
```
